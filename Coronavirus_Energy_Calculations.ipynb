# Binding Energy

```python
sfxn = get_score_function(True)
```

Molecular interactions between proteins and ligand can help us to understand their function. Measure the binding affinity of small molecules can also be used as an important step in drug discovery. Millions of compounds can be virtually posed into proteins and their binding affinities measured. The top ranking (lowest binding energy) molecules can then be passed along the pipeline for biochemical experimentation.

Use the following formula for binding energy:
$$ E_{binding} = E_{complex} - E_{protein} - E_{ligand }$$

Extracting PDBs

Using cleaned versions of the PDB files removes too many residues and produces inaccurate binding energies.

```python
#Human Receptor
ace2 = pyrosetta.pose_from_pdb("1r42.pdb")
#Covid 19 
cov2 = pyrosetta.pose_from_pdb("6vyb.pdb")
ace2cov2 = pyrosetta.pose_from_pdb("6lzg.pdb")
#SARS 2002
sars = pyrosetta.pose_from_pdb("5xlr.pdb")
ace2sars = pyrosetta.pose_from_pdb("3d0g.pdb")
#Bat Coronavirus
bat = pyrosetta.pose_from_pdb("5gyq.pdb")
ace2bat = pyrosetta.pose_from_pdb("batACE2.pdb")
#Pangolin Coronavirus
pang = pyrosetta.pose_from_pdb("pangolin.pdb")
ace2pang = pyrosetta.pose_from_pdb("pangolinACE2.pdb")
```

Covid 19

```python
print("ACE2: %f" % sfxn(ace2))
print("Complex %f" % sfxn(ace2cov2))
print("Cov19: %f" % sfxn(cov2))

binding_energy = sfxn(ace2cov2) - sfxn(ace2) - sfxn(cov2)
print("Binding Energy: %f" % binding_energy)
```

SARS 

```python
print("ACE2: %f" % sfxn(ace2))
print("Complex %f" % sfxn(ace2sars))
print("SARS: %f" % sfxn(sars))

binding_energy = sfxn(ace2sars) - sfxn(ace2) - sfxn(sars)
print("Binding Energy: %f" % binding_energy)
```

Bat Coronavirus

```python
print("ACE2: %f" % sfxn(ace2))
print("Complex %f" % sfxn(ace2bat))
print("Bat Coronvirus: %f" % sfxn(bat))

binding_energy = sfxn(ace2bat) - sfxn(ace2) - sfxn(bat)
print("Binding Energy: %f" % binding_energy)
```

Pangolin Coronavirus

```python
print("ACE2: %f" % sfxn(ace2))
print("Complex %f" % sfxn(ace2pang))
print("Pangolin Coronvirus: %f" % sfxn(pang))

binding_energy = sfxn(ace2pang) - sfxn(ace2) - sfxn(pang)
print("Binding Energy: %f" % binding_energy)
```
